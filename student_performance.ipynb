```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error
 
 
df=pd.read_csv("student_performance_dataset.csv")

X=df[["Past_Exam_Scores",'Attendance_Rate','Study_Hours_per_Week']]
y=df["Final_Exam_Score"]

ln=LinearRegression()

ln.fit(X,y)

test=ln.predict(X)

mean=mean_absolute_error(y,test)
mean_sq=mean_squared_error(y,test)
mean_sq_root=np.sqrt(mean_sq)
print("mean valu is",mean)
print("mean squared is ",mean_sq)
print("mean squared root is ",mean_sq_root)

test1=ln.predict([[90,80,30]])

print("final result",test1)

plt.Figure(figsize=(10,5))
plt.hist(df["Final_Exam_Score"], bins=30,color="skyblue", edgecolor='black')
plt.title("final result preduction")
plt.xlabel("number of student")
plt.ylabel("final result")
plt.grid(True)
plt.show()



fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df["Past_Exam_Scores"], df["Attendance_Rate"], y, color="skyblue", label="Actual Score")
ax.set_xlabel("Past Exam Scores")
ax.set_ylabel("Attendance Rate")
ax.set_zlabel("Final Exam Score")
plt.title("Final Result (3D Scatter)")
plt.legend()
plt.show()
```
